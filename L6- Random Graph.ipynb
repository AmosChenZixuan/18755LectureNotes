# Erdos-Renyi, G(n;p)

An undirected Graph where
    - n: # of nodes
    - p: edge probability between every pair of nodes
    
Between all  $ \binom{n}{2} $ pairs of nodes, there is an edge with prob.p or no edge with prob. 1-p INDEPENDENTLY from each other

## Mean Degree
- Total # of edges in G(n;p) is unfixed since it is random
- It's mean is given by $$
<m> = \binom{n}{2} * p = \frac{n*(n-1)}{2} * p
$$

- degree of a node is $$
<k> = \frac{2*<m>}{n} = (n-1)*p
$$

- probability distribution of degree of a node in G(n;p)
$$
    Pr(k_i = l) = \binom{n-1}{l}*p^l * (1-p)^{n-1-l}
$$
Probability that $V_i$ is connected to $l$ out of n-1 nodes. Known as the Binomial(n-1;p) / (# trials; success pr)

## Degree Variance

$var(k_i) = (n-1)*p*(1-p)$

## Degree Distribution
Poissson \
$Pr(k_i = l) = \frac{e^{-<k>} * <k>^l}{l!}$ for each $l=0,1,2..$


## Clustering Coeffiecent
Still Random\
It's mean value can be captured\
Recall local clustering coeffiecent:\
$C_i$ = # of triangles $v_i$ participates in / $\binom{k_i}{2}$ \
$$E[C_i] = E[ E[C_i | k_i=k] ]$$
$$=E[p]$$
$$=p$$
mean local cc is p. we can also calculate the mean value of average cc:\
$$
E[<C>] = \frac{1}{n}\sum_{i=1}E[C_i]
$$
$$=p$$
